```python

def explore_pdf(file_path):
   
    with pdfplumber.open(file_path) as pdf:
        print(f" Total Pages: {len(pdf.pages)}")
        print(f"{'─'*50}")
        
        for i, page in enumerate(pdf.pages[:3]):  # Pehle 3 pages
            text = page.extract_text()
            print(f"\n Page {i+1} Preview (first 200 chars):")
            print(text[:200] if text else " No text found")
            print(f"{'─'*50}")


sample_text = """
QUARTERLY BUSINESS REPORT - Q4 2024
TechCorp International

Executive Summary:
Revenue increased by 34% compared to Q3, reaching $2.4 billion.
Cloud services division contributed $900 million to total revenue.
Operating expenses rose 12% due to R&D investments.

Key Achievements:
- Launched 3 new AI-powered products
- Customer retention improved to 94%
- Expanded to Singapore and Dubai markets
- Board approved $500M share buyback program

Financial Highlights:
Net profit margin: 23%
EBITDA: $780 million
Cash reserves: $3.2 billion
Debt-to-equity ratio: 0.4

Outlook:
Next quarter targets 40% revenue growth with focus on Asia-Pacific expansion.
New partnerships with 5 Fortune 500 companies expected to close.
"""

print(" Sample Document Stats:")
print(f"  Total characters : {len(sample_text)}")
print(f"  Total words      : {len(sample_text.split())}")
print(f"  Total sentences  : {len([s for s in sample_text.split('.') if s.strip()])}")
print(f"\n Raw Text Preview:")
print(sample_text[:300])
```
